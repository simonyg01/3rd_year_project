```python
States = np.unique(data['State'])
print(States)
```

```python
plt.style.use('ggplot')

data = data[data['Year'] == 2021]

for state in States:
    data1 = data[data['State'] == state]
    #print(data1)
    plt.plot(data1['Month'], data1['COVID-19 Deaths'], 'bo')
    plt.show()
```
